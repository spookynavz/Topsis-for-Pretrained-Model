```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
```

##  Step 1: Create the decision matrix (evaluation of models)

```python
model_names = ['GPT-3', 'GPT-2', 'BERT', 'T5']
evaluation_metrics = ['Performance (BLEU)', 'Speed (Response Time)', 'Memory Consumption (MB)']

# Evaluation evaluation_scores (rows are model_names, columns are evaluation_metrics)
evaluation_scores = np.array([
    [85.3, 0.25, 1200],  # GPT-3
    [82.1, 0.15, 800],   # GPT-2
    [87.6, 0.10, 1000],  # BERT
    [80.4, 0.20, 950],   # T5
])

evaluation_df = pd.DataFrame(evaluation_scores, columns=evaluation_metrics, index=model_names)
```

##  Step 2: Normalize the decision matrix

```python
normalized_evaluation_scores = evaluation_df.copy()
for i in range(len(evaluation_metrics)):
    norm = np.linalg.norm(evaluation_df.iloc[:, i])  # Euclidean norm for each column
    normalized_evaluation_scores.iloc[:, i] = evaluation_df.iloc[:, i] / norm
```

## Step 3: Assign weights (example: equal weights for all criteria)

```python
weights = np.array([1/3, 1/3, 1/3])  # Equal weight for each criterion
```

## Step 4: Calculate the weighted normalized matrix

```python
weighted_evaluation_scores = normalized_evaluation_scores * weights
```

## Step 5: Determine the ideal and negative ideal solutions

```python
ideal_solution = weighted_evaluation_scores.max(axis=0)  # Best for all evaluation_metrics
negative_ideal_solution = weighted_evaluation_scores.min(axis=0)  # Worst for all evaluation_metrics
```

## Step 6: Calculate the separation measures (Euclidean distance)

```python
distance_ideal = np.sqrt(((weighted_evaluation_scores - ideal_solution) ** 2).sum(axis=1))
distance_negative_ideal = np.sqrt(((weighted_evaluation_scores - negative_ideal_solution) ** 2).sum(axis=1))
```

## Step 7: Calculate relative closeness to the ideal solution

```python
closeness_to_ideal = distance_negative_ideal / (distance_ideal + distance_negative_ideal)
```

## Step 8: Rank the models based on closeness

```python
ranking = pd.Series(closeness_to_ideal, index=model_names).sort_values(ascending=False)
```

## Step 9: Visualize the results

```python
# 1. Bar Chart for TOPSIS Rankings
plt.figure(figsize=(10, 6))
plt.bar(ranking.index, ranking.values, color='skyblue')
plt.title("TOPSIS Rankings of Text Generation Models")
plt.xlabel("Models")
plt.ylabel("Relative Closeness to Ideal Solution")
plt.show()
```
